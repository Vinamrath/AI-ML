# 📌 Phase 1 — Problem Definition

## 1️⃣ Define Prediction Unit

### 🔹 What does one row represent?

When you open a dataset, it looks like a table:

| Customer_ID | Age | Income | Bought_Product |
| ----------- | --- | ------ | -------------- |
| 101         | 25  | 50000  | Yes            |
| 102         | 40  | 80000  | No             |

Each **row** is one observation.

That observation could represent:

* One **customer**
* One **house**
* One **transaction**
* One **student**
* One **day**
* One **product**
* etc.

---

## 🔹 What does “Prediction must align with row-level entity” mean?

It means:

👉 **The thing you are predicting must match what one row represents.**

In simple words:

> Your model predicts ONE outcome per row.

---

### ✅ Example 1 — Customer Churn

If one row = one customer
Then prediction = Will THIS customer churn?

✔ Correct alignment.

---

### ❌ Wrong Example

If one row = one transaction
But you try to predict:
"Will this customer churn?"

⚠ Problem:
You have multiple rows per customer (many transactions).
But churn is a **customer-level prediction**, not transaction-level.

👉 Prediction does NOT align with row entity.

---

### ✅ Example 2 — House Price Prediction

If one row = one house
Then prediction = Price of THAT house

✔ Perfect alignment.

---

### ✅ Example 3 — Sales Forecast

If one row = one day
Then prediction = Sales for THAT day

✔ Correct.

---

# 📌 1️⃣ What is **Churn Rate?**

### 🔹 Definition:

> The percentage of customers who leave during a specific time period.

### 🔹 Formula:

[
\text{Churn Rate} = \frac{\text{Customers who left}}{\text{Total customers at start}} \times 100
]

---

### 🔹 Example:

Start of month → 1000 customers
During month → 50 customers cancel

[
Churn = \frac{50}{1000} \times 100 = 5%
]

So churn rate = **5%**

---

# 📌 2️⃣ What is **Retention Rate?**

### 🔹 Definition:

> The percentage of customers who stay.

### 🔹 Formula:

[
\text{Retention Rate} = 100% - \text{Churn Rate}
]

Using previous example:

[
Retention = 100 - 5 = 95%
]

So 95% customers stayed.

# 📌 1️⃣ What is Customer Lifetime Value (CLV)?

### 🔹 Definition:

> CLV = Total money a customer generates for the company during their entire relationship.

In simple words:

👉 “How much is one customer worth to us?”

---

### 🔹 Simple Formula (Basic Version)

[
CLV = \text{Average Revenue per Month} \times \text{Average Customer Lifetime}
]

---

### 🔹 Example

Customer pays $50 per month
Average stay = 24 months

[
CLV = 50 \times 24 = 1200
]

So one customer is worth **$1200**.

---

# 📌 2️⃣ How Churn Affects Revenue (Very Important)

Now think carefully.

If churn increases → customer lifetime decreases.

Let’s see:

---

### Scenario A — Low Churn

Customer stays 24 months
CLV = $1200

---

### Scenario B — High Churn

Customer stays only 12 months
CLV = 50 × 12 = $600

⚠ Company lost $600 per customer.

Now imagine 10,000 customers…

Loss = 600 × 10,000 = **$6,000,000**

That’s why churn is dangerous.

---

# 📌 3️⃣ Mathematical Relationship Between Churn & Lifetime

There’s a powerful formula:

[
\text{Average Lifetime} = \frac{1}{\text{Churn Rate}}
]

If churn rate = 5% (0.05)

[
Lifetime = 1 / 0.05 = 20 \text{ months (approx)}
]

If churn increases to 10% (0.10)

[
Lifetime = 1 / 0.10 = 10 \text{ months}
]

🔥 Churn doubled → lifetime halved.
